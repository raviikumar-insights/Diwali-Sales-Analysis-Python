```python
# import python libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# import csv file 
df = pd.read_csv('Diwali Sales Data.csv',encoding = 'unicode_escape')
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
# drop unrelated/blank columns
df.drop(['Status' ,'unnamed1'], axis=1 , inplace=True) 
```

```python
# check for null values
pd.isnull(df).sum()
```

```python
# drop null values
df.dropna(inplace=True)
```

```python
# change data Types
df['Amount'] = df['Amount'].astype(int)
```

```python
df['Amount'].dtype
```

```python
df.columns
```

```python
# Rename Column
df.rename(columns = {'Marital_Status' : 'Shaadi'})
```

```python
# describe() method returns description of the data in the DataFrame (i.e. count, mean, std, etc)
df.describe()
```

```python
# use describe() for specific columns
df[['Age', 'Orders', 'Amount']].describe()
```

# Exploratory Data Analysis

### Gender

```python
# plotting a bar chart for Gender and it's count
gender_count = df["Gender"].value_counts()

bars = plt.bar(gender_count.index,gender_count.values)
plt.bar_label(bars)

plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Gender Distribution")

plt.xticks(rotation=0)
```

```python
# plotting a bar chart for gender vs total amount
total_amount = df.groupby("Gender")["Amount"].sum().plot(kind='bar')

plt.xticks(rotation=0)
plt.ylabel("Amount")
```

*From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men*

### Age

```python
total_age = df.groupby(["Age Group","Gender"]).size().unstack().plot(kind = 'bar')

plt.ylabel("Count")
plt.xticks(rotation=0)
```

```python
# Total Amount vs Age Group
total_age = df.groupby("Age Group")["Amount"].sum().plot(kind = 'bar')

plt.xticks(rotation=0)
plt.ylabel("Amount")
```

*From above graphs we can see that most of the buyers are of age group between 26-35 yrs female*

### State

```python
data = df.groupby("State")["Orders"].sum()

plt.barh(data.index,data.values)
plt.xlabel("Order")
plt.ylabel("State")
```

```python
data = df.groupby("State")["Amount"].sum()

plt.barh(data.index,data.values)
plt.xlabel("Amount")
plt.ylabel("State")
```

*From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*


### Marital Status

```python
total_marital = df['Marital_Status'].value_counts()
bars = plt.bar(total_marital.index,total_marital.values)

plt.bar_label(bars)
plt.xticks([0,1])

plt.xlabel('Marital_Status')
plt.ylabel('Count')
```

```python
total_amount = df.groupby(['Marital_Status','Gender'])['Amount'].sum().unstack().plot(kind= 'bar')

plt.xticks(rotation=0)
plt.ylabel('Amount')
```

*From above graphs we can see that most of the buyers are married (women) and they have high purchasing power*

### Occupation

```python
total_count = df['Occupation'].value_counts()
plt.figure(figsize=(22,6))
bars = plt.bar(total_count.index,total_count.values)
plt.bar_label(bars)

plt.xlabel('Occupation')
plt.ylabel('Count')
```

```python
total_amount = df.groupby('Occupation')['Amount'].sum()
plt.figure(figsize=(22,6))
bars = plt.bar(total_amount.index,total_amount.values)

plt.xlabel('Occupation')
plt.ylabel('Amount')
```

*From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector*

### Product Category

```python
data = df['Product_Category'].value_counts()
plt.figure(figsize=(10,7))
bars = plt.barh(data.index,data.values)
plt.xlabel('Count')
plt.ylabel('Product_Category')
```

```python
data = df.groupby('Product_Category')['Amount'].sum()
plt.figure(figsize=(10,6))
bars = plt.barh(data.index,data.values)

plt.xlabel('Amount')
plt.ylabel('Product_Category')
```

```python
data = df.groupby('Product_ID')['Orders'].sum().head(10)
plt.figure(figsize=(12,6))
bars = plt.bar(data.index,data.values)

plt.xlabel('Product_ID')
plt.ylabel('Orders')
```
